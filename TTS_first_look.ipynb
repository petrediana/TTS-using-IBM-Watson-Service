```python
from ibm_watson import TextToSpeechV1
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

```python
apikey = 'THE API KEY'
url = 'THE URL'
```

```python
# Setup Service
authenticator = IAMAuthenticator(apikey)
tts = TextToSpeechV1(authenticator=authenticator)
tts.set_service_url(url)
```

```python

with open('./speech.mp3', 'wb') as audio_file:
    res = tts.synthesize('Hello World!', accept='audio/mp3', voice='en-US_AllisonV3Voice').get_result()
    audio_file.write(res.content)
```

```python
import json
voices = tts.list_voices().get_result()
print(json.dumps(voices, indent=2))
```

```python
with open('./test_uk_voice_v3.mp3', 'wb') as audio_file:
    res = tts.synthesize('I am too lazy to read so it is better to listen. Trying out an UK voice over this text that I just typed. Today is Saturday 15th of May and this text to speech is awesome!',
                         accept='audio/mp3', voice='en-GB_KateV3Voice').get_result()
    audio_file.write(res.content)
```

```python
# READING FROM FILE

with open('some_text.txt', 'r') as f:
    text = f.readlines()
    
text = [line.replace('\n','') for line in text]
text = ''.join(str(line) for line in text)

with open('./some_text_to_speech.mp3', 'wb') as audio_file:
    res = tts.synthesize(text, accept='audio/mp3', voice='en-GB_JamesV3Voice').get_result()
    audio_file.write(res.content)
```
